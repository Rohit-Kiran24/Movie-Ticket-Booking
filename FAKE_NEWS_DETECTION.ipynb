```python
import re
```

```python
def wordopt(text):
  text=text.lower()   # convet to lowercase
  text=re.sub(r'https?://\S+\www\.\S+','',text) # remove urls
  text=re.sub(r'<.*?>','',text)  #remive html tags
  text=re.sub(r'[^\w\s]','',text) # remove punctuation
  text=re.sub(r'\d','',text) #digits
  text=re.sub(r'\n',' ',text) # replace newline with spaces
  return text
```

```python
news['text']=news['text'].apply(wordopt)
```

```python
news['text']
```

```python
from sklearn.model_selection import train_test_split
```

```python
x = news['text']
y = news['label']
x_train,x_test,y_train,y_test=train_test_split(x,y,test_size=0.3)
```

```python
from sklearn.feature_extraction.text import TfidfVectorizer
```

```python
vectorization=TfidfVectorizer()
```

```python
xv_train=vectorization.fit_transform(x_train)
```

```python
xv_test=vectorization.transform(x_test)
```

```python
xv_test
```

```python
from sklearn.linear_model import LogisticRegression
```

```python
LR=LogisticRegression()
```

```python
LR.fit(xv_train,y_train)
```

```python
pred_lr=LR.predict(xv_test)
```

```python
LR.score(xv_test, y_test)
```

```python
from sklearn.metrics import classification_report
```

```python
print(classification_report(y_test, pred_lr))
```

```python
def output_lable(n):
    if n == 0:
        return "Fake News"
    elif n == 1:
        return "Genuine news "
```

```python
def manual_testing(news):
    testing_news = {"text": [news]}   # Corrected syntax for defining dictionary
    new_def_test = pd.DataFrame(testing_news)
    new_def_test["text"] = new_def_test["text"].apply(wordopt)
    new_x_test = new_def_test["text"]
    new_xv_test = vectorization.transform(new_x_test)   # Assuming 'vectorization' is your vectorizer object

    pred_lr = LR.predict(new_xv_test)
    # pred_dtc = dtc.predict(new_xv_test)
    pred_gbc = gbc.predict(new_xv_test)
    pred_rfc = rfc.predict(new_xv_test)

    return "\n\nLR Prediction: {} \
            \nGBC Prediction: {} \
            \nRFC Prediction: {}".format(
        output_lable(pred_lr[0]),
        output_lable(pred_gbc[0]),
        output_lable(pred_rfc[0])
    )
```

```python
news_article=str(input())
```
